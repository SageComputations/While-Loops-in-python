# A while loop in any programming language is considered a control flow statement, which allows some code to be executed repeatedly under a specific condition is met.

# A while loop need not necessarily work with sequence but it does perform iterations. 
# At the start of each iteration, it'll check whether a specific condition evaluates to true and only go forward if it is.
# Once the condition evaluates to false, the while loop will terminate.


```python
number = 20

while number < 20:
    print(number)
```

```python
number = 20

while number < 25:
    print(number)
```

# In case of for loops the number of iterations is limited by the number of elements in the list or the sequence over which the for loop is iterating.
# In case of while loop , the iteration is controlled by the condition, and unless the condition changes the while loop can go on and on forever.

# The key to any while loop is to alter the condition so that at some point it will evaluate to false and the while loop will terminate.

```python
number = 20

while number < 25:
    print(number)
    
    number = 26
```

```python
number = 20

while number < 25:
    print(number)
    
    number = 24
```
